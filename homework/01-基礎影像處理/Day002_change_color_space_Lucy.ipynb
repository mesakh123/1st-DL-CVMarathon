- 改變圖片的 color space (`HLS`, `LAB`) 並呈現

```python
import cv2
from matplotlib import pyplot as plt
```

```python
img_path = 'data/lena.png'

# 以彩色圖片的方式載入
img = cv2.imread(img_path, cv2.IMREAD_COLOR)
plt.imshow (img)
plt.show()
# cv2.imshow('ori', img)
# cv2.waitKey(0)
# cv2.destroyAllWindows()

show_img = cv2.cvtColor(img , cv2.COLOR_BGR2RGB ) #COLOR_BGR2RGB:將BGR轉成RGB
plt.imshow (show_img) 

```

```python
# 改變不同的 color space
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV) #COLOR_BGR2HSV
h, s, v = cv2.split(img_hsv)
img2 = cv2.merge([v, s, h])
plt.imshow(img2)
plt.show()

img_hsv2 = cv2.cvtColor(show_img, cv2.COLOR_RGB2HSV) #COLOR_RGB2HSV
plt.imshow (img_hsv2) 
```

```python
#改變圖片的 color space (HLS, LAB) 並呈現
img_hls = cv2.cvtColor(show_img, cv2.COLOR_RGB2HLS)
plt.imshow(img_hls)
```

```python
img_lab = cv2.cvtColor(show_img, cv2.COLOR_RGB2LAB)
plt.imshow(img_lab)
```
